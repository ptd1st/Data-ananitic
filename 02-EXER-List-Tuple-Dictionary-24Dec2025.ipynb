```python
#1. สร้างตัวแปร list ชื่อ animals เพื่อเก็บข้อมูลสัตว์ได้แก่ horse, monkey, cat, dog, tiger, lion, elephant และพิมพ์ออกมา
animals =["horse","monkey","cat","dog","tiger","lion","elephant"]
print(animals)
```

```python
#5. สร้างตัวแปร tuple ชื่อ country ของชื่อประเทศได้แก่ Malaysia, China, Lao, France, Canada, India และพิมพ์ออกมา
#พร้อมทั้งแสดง type ของตัวแปร tuple ดังกล่าว
country = ("Malaysia" ,"China","Lao" ,"France" , "Canada" , "Indai")
print(country, type(country))
```

```python
#6. เพิ่ม Poland เข้าไปใน tuple ที่สร้างในข้อ 5 และพิมพ์ออกมมา
newcountry = list(country)
newcountry.insert(6,"Poland")
print(newcountry)

country = tuple(newcountry)
print(country)
```

```python
#7. สร้างตัวแปรประเภท dictionary ชื่อ car ที่มีปีทีผลิตและชื่อรถยนต์ และพิมพ์ข้อมูลออกมา โดยมีข้อมูลดังนี้
#2001, 2011, 2020 เป็นคีย์ มีค่าที่คู่กันคือ "Toyota", "BMW", Volvo ตามลำดับ 
car = {"2001":"Toyota" , "2011": "BMW" , "2020":"Volvo"}
print(car)
```

```python
#8.จากข้อ 7 ใช้ function pop เพื่อลบ BMW ออก 
# พิมพ์สมาชิกออกมาหลังจากลบข้อมูลออกไปแล้ว
car.pop("2011")
print(car)
```

```python
#9.จากข้อ 7 เปลี่ยนรถยนต์ Toyota เป็น Honda พิมพ์ออกมา
car["2001"] = "Honda"
print(car)
```

```python
#10. เพิ่มข้อมูลในตัวแปร car ปี 2015 รถยนต์ Mazda พิมพ์ข้อมูลออกมา
car["2015"] = "Mazda"
print(car)
```

```python
#13. สร้างตัวแปรที่เป็น set ซึ่งมีสมาชิก  “apple”, “banana”, “mango”, and “orange”.
set = {"apple","banana","mango","orange"}
print(set)
```

```python
#14. สร้างตัวแปร set ชุดที่่1 ที่มีสมาชิก {10, 20, 30, 40, 50} และตัวแปร set ชุดที่ 2 ที่มีสมาชิก {30, 40, 50, 60, 70}
#เชียน python code เพื่อนำข้อมูลสมาชิกของ set ชุดที่ 1 และชุดที่ 2 
#มาใช้เพื่อให้สมาชิกใน set ที่ประกอบไปด้วย {70, 40, 10, 50, 20, 60, 30}
# Hints : ข้อนี้ให้ใช้ union buit-in function
set1 = {10,20,30,40,50}
set2 = {30,40,50,60,70}
newset = set1.union(set2)
print(newset)
```

```python
#15. จากคำตอบข้อ 14. เรียงข้อมูลจากมากไปน้อย  และแสดง data type ของข้อมูลที่เรียงจากมากไปน้อยดังกล่าว
newwset = sorted(newset)
print(newwset)
newwset = sorted(newset,reverse = True)
print(newwset , type(newwset))
```
